```python
import pandas as pd
import datetime
```

```python
tran_df = pd.read_csv('Transactions.csv')
cust_df = pd.read_csv('Customer.csv')
cat_df = pd.read_csv('prod_cat_info.csv')
```

```python
tran_df.head()
```

```python
tran_df.tran_date = pd.to_datetime(tran_df['tran_date'], dayfirst=True)
```

```python
tran_df['prod_cat_code'] = tran_df['prod_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
tran_df['prod_subcat_code'] = tran_df['prod_subcat_code'].astype(str).str.pad(2, side='left', fillchar='0')
```

```python
tran_df.dtypes
```

```python
cust_df.head()
```

```python
cust_df.DOB = pd.to_datetime(cust_df['DOB'])
```

```python
cust_df.dtypes
```

```python
cat_df.head()
```

```python
cat_df['prod_cat_code'] = cat_df['prod_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
cat_df['prod_sub_cat_code'] = cat_df['prod_sub_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
cat_df.head()
```

```python
tran_df.to_csv('Transactions_clean.csv', index=False)
cust_df.to_csv('Customer_clean.csv', index=False)
cat_df.to_csv('prod_cat_info_clean.csv', index=False)
```

```python
total_df = pd.merge(tran_df, cust_df, left_on='cust_id', right_on='customer_Id')
total_df = pd.merge(total_df, cat_df, left_on = ['prod_cat_code', 'prod_subcat_code'], right_on = ['prod_cat_code', 'prod_sub_cat_code'])
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///test.db")
tran_df.to_sql("tran_df", engine, if_exists="replace")
cust_df.to_sql("cust_df", engine, if_exists="replace")
cat_df.to_sql("cat_df", engine, if_exists="replace")
```
