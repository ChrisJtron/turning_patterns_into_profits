```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data= pd.read_csv("../turning_patterns_into_profits/Data/Transactions_all_data.csv")

data.head()
```

```python
data.drop(["Unnamed: 15", "Unnamed: 16", "Unnamed: 17", "Unnamed: 18", "Unnamed: 19", "Unnamed: 20"], axis=1, inplace=True)
```

```python
month_list=[]
year_list=[]
mo_yr_list=[]

for each_date in data['tran_date']:
    date= datetime.datetime.strptime(str(each_date), "%m/%d/%Y")
    month_list.append(date.month)
    year_list.append(date.year)
    mo_yr_list.append(f"{date.month}/{date.year}")
```

```python
data["transaction_month"] = month_list
data["transaction_year"] = year_list
data["transaction_MoYr"] = mo_yr_list

data.head()
```

##  E-Shop sales by Category
##  all of Category 1 is increasing
##  3-4 and 3-5 are declining
##  3-9 is steady.  3-10 is increasing
##  5-11 is declining

```python
## create df to analyze category 1

edata1=data.loc[(data['Store_type']=='e-Shop') & (data['prod_cat_code']==1)]
edata1.head()
```

```python
edata1_1=edata1.loc[edata1['prod_subcat_code']==1]
edata1_1=edata1_1.groupby('transaction_year').sum()
edata1_1=edata1_1.reset_index()
edata1_1=edata1_1.drop(3)
edata1_1
```

```python
## Category 1

subcat=[1,3,4]

for each in subcat:
    edata1x=edata1.loc[edata1['prod_subcat_code']==each]
    edata1x=edata1x.groupby('transaction_year').sum()
    edata1x=edata1x.reset_index()
    edata1x=edata1x.drop(3)
    edata1x
    
    X=edata1x["transaction_year"].values.reshape(-1,1)
    y=edata1x["total_amt"].values.reshape(-1,1)
    
    plt.scatter(X,y)
    plt.title(f"E-Shop Category 1, Subcategory {each}")
    
    xx=edata1x["transaction_year"]
    yy= edata1x["total_amt"]
    
    m,b=np.polyfit(xx, yy,1)
    plt.plot(xx, m*xx + b)
    
    plt.show()
    
    model=LinearRegression()
    model.fit(X,y)
    Score=model.score(X,y)
    prediction=model.predict([[2014]])
    
    predicted=model.predict(X)

    mse=mean_squared_error(y,predicted)
    
    print(f"Model Score: {Score}    MSE: {mse}    2014 Prediction: {prediction}")
```

```python
## create df to analyze category 3

edata3=data.loc[(data['Store_type']=='e-Shop') & (data['prod_cat_code']==3)]
edata3.head()
```

```python
##Category 3

subcat=[4,5,9,10]

for each in subcat:
    edata3x=edata3.loc[edata3['prod_subcat_code']==each]
    edata3x=edata3x.groupby('transaction_year').sum()
    edata3x=edata3x.reset_index()
    edata3x=edata3x.drop(3)
    edata3x
    
    X=edata3x["transaction_year"].values.reshape(-1,1)
    y=edata3x["total_amt"].values.reshape(-1,1)
    
    plt.scatter(X,y)
    plt.ylim(200000,400000)
    plt.title(f"E-Shop Category 3, Subcategory {each}")
    
    
    xx=edata3x["transaction_year"]
    yy= edata3x["total_amt"]
    
    m,b=np.polyfit(xx,yy,1)
    plt.plot(xx, m*xx + b)
    
    plt.show()
    
    model=LinearRegression()
    model.fit(X,y)
    Score=model.score(X,y)
    prediction=model.predict([[2014]])
    
    predicted=model.predict(X)

    mse=mean_squared_error(y,predicted)
    
    print(f"Model Score: {Score}    MSE: {mse}      2014 Prediction: {prediction}")
```
