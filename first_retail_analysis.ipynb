```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('../turning_patterns_into_profits/Data/Transactions_all_data.csv')

data.head()
```

```python
import datetime

month_list=[]
year_list=[]
mo_yr_list=[]

for each_date in data['tran_date']:
    date= datetime.datetime.strptime(str(each_date), "%m/%d/%Y")
    month_list.append(date.month)
    year_list.append(date.year)
    mo_yr_list.append(f"{date.month}/{date.year}")
```

```python
# Checking to make sure new lists are same length as dataframe columns

print(f"month: {len(month_list)}")
print(f"year: {len(year_list)}")
print(f"prod_cat: {len(data['prod_cat'])}")
print(mo_yr_list)
```

```python
data["transaction_month"] = month_list
data["transaction_year"] = year_list
data["transaction_MoYr"] = mo_yr_list

data.head()
```

```python
data.drop(["Unnamed: 15", 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20'], axis=1)
```
